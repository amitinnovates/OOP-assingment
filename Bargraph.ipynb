```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt

components = ['CPU', 'GPU', 'RAM', 'Storage', 'Motherboard']
percentages = [30, 25, 20, 15, 10]

plt.pie(percentages, labels=components, autopct='%1.1f%%', startangle=90)
plt.title("Daily Selling Percentage of Hardware Components")
plt.axis('equal')
plt.show()
```
